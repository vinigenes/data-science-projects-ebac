```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Configuração visual
sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (10, 6)

# Leitura do arquivo CSV
df = pd.read_csv("/content/ecommerce_preparados.csv")

# Visualização inicial
df.head()
```

1. Gráfico de Histograma
- Distribuição dos preços dos produtos
  
  Objetivo: entender concentração de preços e possíveis outliers.

```python
plt.hist(df["Preço"].dropna(), bins=30)
plt.title("Distribuição dos Preços dos Produtos")
plt.xlabel("Preço")
plt.ylabel("Frequência")
plt.show()
```

2. Gráfico de Dispersão
- Relação entre preço e nota do produto
  
  Objetivo: verificar se produtos mais caros tendem a ter melhores avaliações.

```python
plt.scatter(df["Preço"], df["Nota"])
plt.title("Relação entre Preço e Nota")
plt.xlabel("Preço")
plt.ylabel("Nota")
plt.show()
```

3. Mapa de Calor
- Correlação entre variáveis numéricas

  Objetivo: identificar relações fortes ou fracas entre métricas do e-commerce.

```python
num_cols = [
    "Preço", "Nota", "N_Avaliações", "Desconto",
    "Preço_MinMax", "Nota_MinMax", "N_Avaliações_MinMax"
]

corr = df[num_cols].corr()

sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Mapa de Calor – Correlação entre Variáveis Numéricas")
plt.show()

```

4. Gráfico de Barras
- Quantidade de produtos por temporada

  Objetivo: entender a distribuição de produtos ao longo das temporadas.

```python
def padronizar_temporada(valor):
    valor = str(valor).lower()

    if "primavera" in valor and "verão" in valor and ("outono" in valor or "inverno" in valor):
        return "Todas as Estações"
    elif "primavera" in valor or "verão" in valor:
        return "Primavera/Verão"
    elif "outono" in valor or "inverno" in valor:
        return "Outono/Inverno"
    elif "não definido" in valor:
        return "Não definido"
    else:
        return "Não definido"

df["Temporada_Padronizada"] = df["Temporada"].apply(padronizar_temporada)

df["Temporada_Padronizada"].value_counts().plot(kind="bar")
plt.title("Distribuição de Produtos por Temporada")
plt.xlabel("Temporada")
plt.ylabel("Quantidade de Produtos")
plt.show()
```

5. Gráfico de Pizza
- Proporção de produtos por gênero

  Objetivo: visualizar a participação de cada gênero no catálogo.

```python
def padronizar_genero(valor):
    valor = str(valor).lower()

    if "femin" in valor or "mulher" in valor:
        return "Feminino"
    elif "mascul" in valor or "homem" in valor:
        return "Masculino"
    elif "beb" in valor:
        return "Bebês"
    elif "menina" in valor:
        return "Meninas"
    elif "menino" in valor:
        return "Meninos"
    elif "unisex" in valor:
        return "Unissex"
    elif "sem genero" in valor:
        return "Sem gênero"
    else:
        return "Outros"

df["Genero_Padronizado"] = df["Gênero"].apply(padronizar_genero)

genero_counts = df["Genero_Padronizado"].value_counts()

plt.figure(figsize=(8, 8))
plt.pie(
    genero_counts,
    labels=genero_counts.index,
    autopct="%1.1f%%",
    startangle=90
)
plt.title("Distribuição de Produtos por Público-Alvo")
plt.axis("equal")
plt.show()


```

Gráfico de Densidade
- Densidade das notas dos produtos

  Objetivo: observar concentração e dispersão das avaliações.

```python
sns.kdeplot(df["Nota"].dropna(), fill=True)
plt.title("Densidade das Notas dos Produtos")
plt.xlabel("Nota")
plt.ylabel("Densidade")
plt.show()
```

Gráfico de Regressão
- Relação entre preço e número de avaliações

  Objetivo: analisar se produtos mais caros recebem mais ou menos avaliações.

```python
sns.regplot(x="Preço", y="N_Avaliações", data=df)
plt.title("Regressão: Preço vs Número de Avaliações")
plt.xlabel("Preço")
plt.ylabel("Número de Avaliações")
plt.show()
```
