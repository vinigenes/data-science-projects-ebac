Abaixo temos um dataframe com as seguintes informações:


**Nome:** O nome de cada indivíduo da base


**Idade:** A idade de cada um deles


**Profissão:** A atual ocupação


**Salário:** O salário registrado


**Limite de crédito:** O limite de crédito disponível para cada um dos indivíduos


**Historico_Inadimplencia:** Marco como 0 aqueles que nunca tiveram um histórico de inadimplência e como 1 aqueles que já tiveram em algum momento.


**Estado civil:** O estado civil de cada um deles


**Imovel_Proprio:** Marcado como 0 aqueles que não tem imóvel próprio e como 1 aqueles que tem.

```python
import pandas as pd
```

```python

data = {
    'Nome': ['Alice', 'Joao', 'Charlie', 'David', 'Eva', 'Diego', 'Denize', 'Claudio'],
    'Idade': [25, 30, 35, 40, 45, 60, 22, 24],
    'Profissão': ['Engenheiro', 'Médico', 'Professor', 'Advogado', 'Médico','Engenheiro', 'Estudante','Estudante'],
    'Salário': ['4500', '8000', '5000', '10000', '12000','15000', '1200','1500'],
    'Limite_Credito': ['2500', '4000', '4000', '1000', '10000','2000', '500','250'],
    'Historico_Inadimplencia': ['0', '0', '0', '1', '0','1', '0','1'],
    'Estado_Civil': ['Casamento', 'Casamento', 'Solteiro', 'Solteiro', 'Casamento','Solteiro', 'Solteiro','Solteiro'],
    'Imovel_Proprio': ['0', '0', '0', '1', '1','1', '0','0']
}

df = pd.DataFrame(data)
df.head(8)
```

A proposta da atividade é que utilizem os dados do DataFrame Data e escolham apenas 3 gráficos que ajudem vocês a identificar quais fatores são os principais impulsionadores para aqueles que tem limite de crédito alto.

Durante a exploração de vocês podem utilizar mais de 3 gráficos mas deixem na entrega final apenas os 3 mais assertivos que contenham a informação da mensagem que querem passar, fiquem livres para utilizar as bibliotecas que quiserem para os gráficos.

Justifiquem a escolha dos visuais e despertem o lado analítico de vocês com uma análise acerca do descoberto através dos gráficos. Lembrem o que aprendemos na aula, devemos ser assertivos na nossa mensagem.

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(6,4))

sns.scatterplot(
    data=df,
    x='Salário',
    y='Limite_Credito'
)

plt.title("Relação entre Salário e Limite de Crédito")
plt.xlabel("Salário")
plt.ylabel("Limite de Crédito")

plt.show()
```

# **O que o gráfico mostra**

Esse gráfico permite observar se pessoas com salários maiores tendem a possuir limites de crédito maiores. Cada ponto representa um indivíduo da base de dados, permitindo visualizar a distribuição entre renda e limite de crédito.

# **Justificativa**

O gráfico mostra a relação entre salário e limite de crédito disponível. Observa-se que indivíduos com salários mais altos tendem a apresentar limites de crédito maiores. Isso sugere que a renda é um fator importante na definição do limite de crédito, pois instituições financeiras consideram a capacidade de pagamento do cliente ao conceder crédito.

```python
df['Historico_Inadimplencia'] = df['Historico_Inadimplencia'].astype(int)

plt.figure(figsize=(6,4))

sns.barplot(
    data=df,
    x='Historico_Inadimplencia',
    y='Limite_Credito',
    errorbar=None
)

plt.title("Limite Médio de Crédito vs Histórico de Inadimplência")
plt.xlabel("Histórico de Inadimplência (0 = Não, 1 = Sim)")
plt.ylabel("Limite Médio de Crédito")

plt.show()
```

# **O que o gráfico mostra**

Ele compara o limite médio de crédito entre pessoas com e sem histórico de inadimplência.

# **Justificativa**

O gráfico mostra a relação entre histórico de inadimplência e limite de crédito. Observa-se que indivíduos sem histórico de inadimplência tendem a possuir limites de crédito maiores. Isso ocorre porque instituições financeiras consideram o histórico de pagamento do cliente ao definir o limite disponível, reduzindo o crédito concedido a clientes considerados de maior risco.

```python
plt.figure(figsize=(6,4))

sns.scatterplot(
    data=df,
    x='Idade',
    y='Limite_Credito',
    hue='Historico_Inadimplencia',
    s=100
)

plt.title("Limite de Crédito vs Idade e Histórico de Inadimplência")
plt.xlabel("Idade")
plt.ylabel("Limite de Crédito")
plt.legend(title="Histórico de Inadimplência")

plt.show()
```

# **O que esse gráfico mostra**

Esse gráfico permite analisar três variáveis ao mesmo tempo:

Idade

*   Idade
*   Limite de crédito
* Histórico de inadimplência

Cada ponto representa um indivíduo.

# **Justificativa**

O gráfico apresenta a relação entre idade, limite de crédito e histórico de inadimplência. Observa-se que indivíduos sem histórico de inadimplência tendem a possuir limites de crédito maiores em diferentes faixas de idade. Além disso, pessoas com maior idade podem apresentar limites mais elevados, possivelmente devido a maior estabilidade financeira e histórico de crédito mais longo. Por outro lado, indivíduos com histórico de inadimplência apresentam limites menores, indicando que esse fator é considerado pelas instituições financeiras ao definir o limite de crédito.
